# KPI and Forecast Exploration

This notebook explores the generated KPIs and demonstrates how to query them for analysis.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../inventory_v2.db")
```

```python
# Query Channel KPIs
query = """
SELECT 
    channel_id,
    total_revenue,
    order_count,
    avg_order_value
FROM channel_kpis
ORDER BY total_revenue DESC
"""

try:
    df = pd.read_sql(query, conn)
    print("Channel KPIs:")
    display(df)
except Exception as e:
    print("channel_kpis table not found. Run the KPI job first.")
```
